```python
# Read the file 
df = pd.read_csv('netflix.csv')
df.head()
```

```python
# Explore file
df.info()
```

```python
# Count null rows 
df.isnull().sum()
```

```python
# Null processing
# Fill Unknown for director, country and cast columns
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# Fill the most appeared values for rating column
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])

# Delete null rows for date_added and release_year
df.dropna(subset=['date_added', 'duration'], inplace=True)

# Check null
df.isnull().sum()
```

```python
# Split the genre collumn
genre_df = df[['country','show_id', 'listed_in']].dropna()

# Split string by ','
genre_df['listed_in'] = genre_df['listed_in'].str.split(',')
df_exploded = genre_df.explode('listed_in')
df_exploded['listed_in'] = df_exploded['listed_in'].str.strip()
```
